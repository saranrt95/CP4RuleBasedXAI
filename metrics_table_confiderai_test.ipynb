```python
import pandas as pd

metrics_all = []
res_main_dir = "results_20250206"
for scenario in ['easy','medium','hard']:
    for rel in ['True', 'False']:
        for sim in ['True', 'False']:
            res_path = f"{res_main_dir}/{scenario}_Rel{rel}_globalSim{sim}/metrics.csv"
            data = pd.read_csv(res_path)
            #data_confiderai = data[data["score"]=="confiderai+"]
            data["scenario"] = scenario 
            data["consider_relevance"] = rel
            data["consider_similarity"] = sim
            metrics_all.append(data)
metric_all_table = pd.concat(metrics_all).reset_index(drop = True)
metric_all_table
```

```python
metric_all_table.to_excel("confiderai_plus_new_metrics.xlsx", index = False)
```
